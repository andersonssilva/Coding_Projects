###  5 - Removing duplicates
If you explore the Google Play data set long enough, you'll notice some apps have duplicate entries. For instance, Instagram has four entries:

```python
for row in dataG[1:]:
    name=row[0]
    if name == 'Instagram':
        print(row)
```

Analyzing the number of duplicates

```python
def dups(data_dups, namecol):
    duplicate_apps = []
    unique_apps = []
    #identifying dups
    for row in data_dups[1:]:
        name = row[namecol]
        if name in unique_apps:
            duplicate_apps.append(name)
        else:
            unique_apps.append(name)
    
    #counting dups by apps
    dups={}
    for row in duplicate_apps:
        if row in dups:
            dups[row]+=1
        else:
            dups[row]=1

    # printing the dups
    print('See the list of duplicated apps below as well as the quantity of repetition:')
    for row in dups:
        print(row, ':',str(dups[row]))
    return unique_apps

ios_unique=[]
android_unique=[]

ios_unique=dups(dataA, 2)
android_unique=dups(dataG, 0)
```

Removing the duplicates found based in the number of reviews; the higher the number of reviews, the more recent the data should be

```python

def clean_data(unique_data, data_d, nameCol, reviewCol):
    clean = []
    clean.append(data_d[0]) #adding header to the clean list
    # creating the dictionary with distinct names of the duplicate apps
    dups_apps={}

    for row in unique_data:
        if row not in dups_apps:
            dups_apps[row]=0
    #print(len(dups_apps))

    # loading the highest review per duplicate apps in the dictionary
    for app in data_d[1:]:
        name = app[nameCol]
        Nreviews = float(app[reviewCol])#G is 3
        if name in dups_apps and dups_apps[name] < Nreviews:
            dups_apps[name] = Nreviews
        elif name not in dups_apps:
            dups_apps[name] = Nreviews
    
    already_added = [] #to register apps already added because we have dups of apps with same name and Nreviews 
    for app in data_d[1:]:
        name = app[nameCol]
        Nreviews = float(app[reviewCol])
        if name in dups_apps and dups_apps[name] == Nreviews and (name not in already_added):
            clean.append(app)
            already_added.append(name)

    print(len(clean))
    return clean

android_clean = []
ios_clean = []

ios_clean = clean_data(ios_unique, dataA, 2, 6)

android_clean = clean_data(android_unique, dataG, 0, 3)
```

Behind the scenes, each character we use in a string has a corresponding number associated with it. For instance, the corresponding number for character 'a' is 97, character 'A' is 65, and character 'Áà±' is 29,233. We can get the corresponding number of each character using the ord() built-in function.

The numbers corresponding to the characters we commonly use in an English text are all in the range 0 to 127, according to the ASCII (American Standard Code for Information Interchange) system. Based on this number range, we can build a function that detects whether a character belongs to the set of common English characters or not. If the number is equal to or less than 127, then the character belongs to the set of common English characters.

All English apps with up to three emoji or other special characters will still be labeled as English.

```python
def is_english(string):
    no_eng=0
    for char in string:
        if ord(char)>127:
            no_eng+=1
    if no_eng > 3:
        return False
    else:
        return True

#is_english('Docs To Go‚Ñ¢ Free Office Suite')
#is_english('Instachat üòú')
is_english('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠')
#is_english('Instagram')
```

```python
android_eng = []
ios_eng = []

def eng_filter(data, name_col_number):
    only_eng = []
    only_eng.append(data[0])
    for row in data[1:]:
        appname=row[name_col_number]
        eng=is_english(appname)
        if eng:
            only_eng.append(row)
    print(len(only_eng))
    return only_eng

android_eng=eng_filter(android_clean, 0)
ios_eng=eng_filter(ios_clean, 2)
```

### 7 - Selecting only free apps

```python
free_apps=[]
def only_free(data_cost, priceCol):
    free_apps=[]
    free_apps.append(data_cost[0]) #adding header to the free apps list
    for row in data_cost[1:]:
        price_ini = row[priceCol]
        dollar=''
        #cleaning $ sign
        for char in price_ini:
            first_char=char[0]
            if first_char == '$':
                dollar = '$'
        if dollar == '$':
            price_final=float(price_ini[1:])
        else:
            price_final=float(price_ini)
       
        #appending free apps
        if price_final == 0.0:
            free_apps.append(row)
    print(len(free_apps))
    return free_apps

ios_free = []
android_free = []

ios_free = only_free(ios_eng, 5)
android_free = only_free(android_eng, 7)

```

### 8 - Creating frequency tables for the kind of apps in both stores

As we mentioned in the introduction, our aim is to determine the kinds of apps that are likely to attract more users because our revenue is highly influenced by the number of people using our apps.

To minimize risks and overhead, our validation strategy for an app idea is comprised of three steps:

Build a minimal Android version of the app, and add it to Google Play.
If the app has a good response from users, we develop it further.
If the app is profitable after six months, we build an iOS version of the app and add it to the App Store.
Because our end goal is to add the app on both Google Play and the App Store, we need to find app profiles that are successful on both markets. For instance, a profile that works well for both markets might be a productivity app that makes use of gamification.

We'll build two functions we can use to analyze the frequency tables:

- One function to generate frequency tables that show percentages

- Another function that we can use to display the percentages in a descending order

```python

#calculate the percentage of the type over the total
def percent_type(free_data, typecol):
    freq = {}
    for row in free_data[1:]:
        apptype = row[typecol]
        if apptype in freq:
            freq[apptype]+=1
        else:
            freq[apptype]=1
    for row in freq:
        percent=(freq[row]/len(free_data))*100
        freq[row]=percent
    return freq


#print(sorted(ios_freq.values(), reverse=True))

#display in descending order creating a list of lists attempt, it works but not exactly the style we are looking for
def desc(freq_data):
    sorted_desc=sorted(freq_data.items(), key=lambda t:t[1], reverse=True)
    print(sorted_desc) 

#display in descending
def display_table(dataset, index):
    table = percent_type(dataset, index)
    table_display = []
    for key in table:
        #invert the order of elements (x%: Category) so we can use the sorted function and order by the key
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
        
    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])#bring back to the correct order (Category:x%)
```

```python

#calculate count of the each type
def rating_count(free_data, typecol, rating_tot_col):
    freq = {}
    for row in free_data[1:]:
        apptype = row[typecol]
        if apptype in freq:
            freq[apptype]+=1
        else:
            freq[apptype]=1
    genre_rating_count={}
    for row in freq:
        total=0
        len_genre=0
        for i in free_data[1:]:
            appgenre=i[typecol]
            if appgenre==row:
                total=total+int(i[rating_tot_col])
                len_genre += 1
        avg_rating_count=total/len_genre
        print(row, ":", avg_rating_count)   
rating_count(ios_free, -5, 6)
```

### Most Popular Apps by Genre on Google Play
For the Google Play market, we actually have data about the number of installs, so we should be able to get a clearer picture about genre popularity. However, the install numbers don't seem precise enough ‚Äî we can see that most values are open-ended (100+, 1,000+, 5,000+, etc.):

```python
display_table(android_free, 5) # the Installs columns
```

One problem with this data is that is not precise. For instance, we don't know whether an app with 100,000+ installs has 100,000 installs, 200,000, or 350,000. However, we don't need very precise data for our purposes ‚Äî we only want to get an idea which app genres attract the most users, and we don't need perfect precision with respect to the number of users.

We're going to leave the numbers as they are, which means that we'll consider that an app with 100,000+ installs has 100,000 installs, and an app with 1,000,000+ installs has 1,000,000 installs, and so on.

To perform computations, however, we'll need to convert each install number to float ‚Äî this means that we need to remove the commas and the plus characters, otherwise the conversion will fail and raise an error. We'll do this directly in the loop below, where we also compute the average number of installs for each genre (category).

```python
#calculate count of the each type
freq = {}
for row in android_free[1:]:
    apptype = row[1]
    if apptype in freq:
        freq[apptype]+=1
    else:
         freq[apptype]=1
for category in freq:
    total = 0
    len_category = 0
    for app in android_free:
        category_app = app[1]
        if category_app == category:            
            n_installs = app[5]
            n_installs = n_installs.replace(',', '')
            n_installs = n_installs.replace('+', '')
            total += float(n_installs)
            len_category += 1
    avg_n_installs = total / len_category
    print(category, ':', avg_n_installs)
```
